```python
def lin_diff():
  a=int(input("Enter Start Value:"))
  b=int(input("Enter End Value:"))
  c=int(input("Enter Number of Intervals:"))
  d=np.linspace(a,b,c)
  e=[]
  for i in range(len(d) - 1):
    e.append(d[i+1] - d[i])
  return e

lin_diff()
```

```python
def lin_diff(a,b,c):
  d=np.linspace(a,b,c)
  e=[]
  for i in range(len(d) - 1):
    e.append(d[i+1] - d[i])
  return e

lin_diff(2,9,8)
```
